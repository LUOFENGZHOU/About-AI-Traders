```python
import numpy as np
import pandas as pd
rolling_days = 5
# data_file = 'acou_and_factors.csv'
data_file = "08053before_rolling.csv"
x_column = ["factor"+str(i) for i in range(0,17)]
for i in range(12):
    x_column.append("10M" + str(i))
target = ['NextReturnCate']
```

```python
all_data = pd.read_csv(data_file)
all_stock_code =list(set(all_data['SecuCode']))
all_rolling_data = []   
index = 0
for code in all_stock_code:
    p = str(index) + '/' + str(len(all_stock_code))
    print(p,end='\r')
    index += 1
    one_stock = all_data.loc[all_data['SecuCode'] == code]
    one_stock = one_stock.sort_values(by=['TradingDay'])
    for j in range(one_stock.shape[0] - rolling_days):
        one_stock_rolling = one_stock.iloc[j: j+rolling_days][x_column].reset_index(drop=True).unstack().to_frame().sort_index(level=1).T
        one_stock_rolling['TradingDay'] = one_stock.iloc[j + rolling_days-1]['TradingDay']
        one_stock_rolling['SecuCode'] = code
        one_stock_rolling['NextReturnCate'] = one_stock.iloc[j + rolling_days-1]['NextReturnCate']
        one_stock_rolling["NextReturn"] = one_stock.iloc[j + rolling_days - 1]["NextReturn"]
        all_rolling_data.append(one_stock_rolling)
#all_rolling_data = pd.concat(all_rolling_data)
```

```python
all_rolling_data = pd.concat(all_rolling_data)
```

```python
b = all_rolling_data
```

```python
b.columns = [col[0]+'_'+str(col[1]) if col[1] != '' else col[0] for col in b.columns.values]
```

```python
b.to_csv('0805rolling5.csv')
```
