```python
x = 'L1_STOCK_MIN'
data_loader = BTMongo(config=my_config, name=x)
y_list = [ _ for _ in data_loader.collist if _[-3:] == 'M10' and (_[:4] == '2017' or _[:4] == '2018' or _[:4] == "2016" or _[:4] == "2015")]
y_list.sort(key=date_to_int)
symbol_list = list(pd.read_csv('all_data.csv')['SecuCode'])
symbol_list = list(set(symbol_list))
symbol_list.sort()
symbol_list = [(6-len(str(i)))*'0' + str(i) for i in symbol_list]
acoustic_feature = [] 
for i in range(len(y_list)):
    for j in range(len(symbol_list)):
        data_roll = []
        data_roll.append(data_loader.get_data(y_list[i],symbol_list[j]))
        y = pd.concat(data_roll)
        if not y.empty:
            y = np.array(y['ClosePrice'])
            yf=abs(fft(y)) 
            columns = ["10M"+str(i) for i in range(int(len(yf)/2))]
            yf = pd.DataFrame([yf[:int(len(yf)/2)]],columns=columns)
            yf['TradingDay'] = pd.to_datetime(y_list[i][:-4])
            yf['SecuCode'] = symbol_list[j]
            acoustic_feature.append(yf)
acoustic_feature = pd.concat(acoustic_feature)
# acoustic_feature
```

```python
# acoustic_feature.to_csv('acoustic_feature.csv')
acoustic_feature.to_csv('acoustic_feature2.csv')
```

```python
all_data = pd.read_csv('all_data.csv')
all_data['TradingDay'] = all_data['TradingDay'].astype('datetime64[ns]')
acoustic_feature['SecuCode'] = acoustic_feature['SecuCode'].astype('int')
```
